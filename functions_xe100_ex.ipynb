```python
c61u = np.linspace(10**-10,10**-8,25)
chi2 = []
for i in range(len(c61u)):
    chi2 += [pfxe.chi2(c61u[i],ER,0)]
    #print(chi2[i])
chi2 = np.asarray(chi2)
```

```python
fig, axs = plt.subplots(1,figsize=(8,6))
axs.set_title('Chi-squared for Xenon100 - C61u: $m_\chi = 10$ GeV',size=20)
axs.set_xlabel('C61u',size=16)
axs.set_ylabel('$\chi^2$',size=16)
axs.plot(c61u,chi2)
axs.tick_params(labelsize=16)
```

```python
init = np.linspace(10**-9,10**-8,10)
c61ubest = []
for i in range(len(init)):
    chimin = pfxe.chi2min(ER,0,init[i],method='nelder-mead')
    c61ubest += [chimin.x[0]]
    #print(chimin.x[0],chimin.fun)
c61ubest = np.asarray(c61ubest)
plt.plot(init,c61ubest)
```
